```python
from requests_html import HTMLSession
import pickle
import json
import os
```

```python
# url = "https://towardsdatascience.com/"
url = "https://dariusforoux.medium.com/"

session = HTMLSession()
r1 = session.get(url)
for s in r1.html.find("script"):
    if "window.__PRELOADED_STATE__" in s.text:
        script = s
preload_data = json.loads(script.text.split("window.__PRELOADED_STATE__ = ")[1])
userid = preload_data['client']['routingEntity']['id']
userid
```

```python
with open("payload", 'rb') as f:
    payload = pickle.load(f)
```

```python
# variables = {"homepagePostsFrom":"L1509878893000", # the time
#                 "includeDistributedResponses":True,
#                 "id":"54163a1723f6", # user id
#                 "username":None,
#                 "homepagePostsLimit":25,
#                 'pagingInfo': {'next': {'from': 'L1478512862320', 'limit': 25, '__typename': 'PageParams'}, '__typename': 'Paging'}, 
#                 '__typename': 'PostConnection'}
variables = {"homepagePostsFrom":"L1509878893000", # the time
                "includeDistributedResponses":True,
                "id":userid, # user id
                "username":None,
                "homepagePostsLimit":25}
```

```python
payload[0]['variables'] = variables
```

```python
r = session.post(os.path.join(url, "_", "graphql"), json=payload)
```
